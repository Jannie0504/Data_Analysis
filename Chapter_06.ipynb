## 2. 조건에 맞는 데이터만 추출하기
### pandas의 query() 이용
#### query('조건') : 조건에 해당하는 행만 추출

```python
import pandas as pd
exam = pd.read_csv('exam.csv')
exam
```

```python
# exam에서 nclass가 1인 경우만 추출
exam.query('nclass == 1')
```

```python
# 2반인 경우만 추출
exam.query('nclass == 2')
```

```python
# 1반이 아닌 경우 추출
exam.query('nclass != 1')
```

```python
# 3반이 아닌 경우 추출
exam.query('nclass != 3')
```

### 초과, 미만, 이상, 이하 조건 걸기

```python
# 수학 점수가 50점을 초과한 경우만 추출
exam.query('math > 50')
```

```python
# 수학 점수가 50점 미만인 경우만 추출
exam.query('math < 50')
```

```python
# 영어 점수가 50점 이상인 경우만 추출
exam.query('english >= 50')
```

```python
# 영어 점수가 80점 이하인 경우만 추출
exam.query('english <= 80')
```

### 여러 조건을 충족하는 행 추출하기
#### & : 그리고, and를 의미

```python
# 1반이면서 수학 점수가 50점 이상인 경우 추출
exam.query('nclass == 1 & math >= 50')
```

```python
# 2반이면서 영어 점수가 80점 이상인 경우 추출
exam.query('nclass == 2 & english >= 80')
```

### 여러 조건 중 하나 이상 충족하는 행 추출하기
#### | : 또는, or을 의미

```python
# 수학 점수가 90점 이상이거나 영어 점수가 90점 이상인 경우 추출
exam.query('math >= 90 | english >= 90')
```

```python
# 영어 점수가 90점 미만이거나 과학 점수가 50점 미만인 경우 추출
exam.query('english < 90 | science < 50')
```

### 목록에 해당하는 행 추출하기
#### in과 []를 이용
#### 변수 이름 in [변수 값1, 변수 값2, ...]

```python
# 1, 3, 5반에 해당하면 추출
exam.query('nclass == 1 | nclass == 3 | nclass == 5')
```

```python
# 1, 3, 5반에 해당하면 추출
exam.query('nclass in [1, 3, 5]')
```

### 추출한 행으로 데이터 만들기

```python
# nclass가 1인 행을 추출해 nclass1에 할당
nclass1 = exam.query('nclass == 1')

# nclass가 2인 행을 추출해 nclass2에 할당
nclass2 = exam.query('nclass == 2')
```

### 문자 변수를 이용해 조건에 맞는 형 추출하기

## 5. 파생변수 추가하기
### assign() 이용
#### assign(파생변수명 = 파생변수를 만드는 공식)

```python
# total 변수 추가하기
exam.assign(total = exam['math'] + exam['english'] + exam['science'])
```

```python
# 여러 파생변수 한 번에 추가하기
exam.assign(total = exam['math'] + exam['english'] + exam['science'],
            mean = (exam['math'] + exam['english'] + exam['science']) / 3)
```

### lambda 이용해 데이터 프레임명 줄여 쓰기

```python
# 긴 데이터 프레임명 지정
long_name = pd.read_csv('exam.csv')

# long_name 직접 입력
long_name.assign(new = long_name['math'] + exam['english'] + exam['science'])

# long_name 대신 x 입력
long_name.assign(new = lambda x : x['math'] + x['english'] + x['science'])
```

## 6. 집단별로 요약하기
### 집단별로 요약하기
### agg() 이용 : 요약 통계량
#### agg(요약값을 할당할 변수명 = (값을 요약하는데 사용할 변수와 함수))

```python
# math 평균 구하기
exam.agg(mean_math = ('math', 'mean'))
```

### groupby() 이용 : 변수의 범주별로 데이터를 분리
#### groupby('변수명')

```python
# nclass 별로 분리, math 평균 구하기
exam.groupby('nclass').agg(mean_math = ('math', 'mean'))
```

```python
# 변수를 인덱스로 바꾸지 않기
# groupby()의 as_index 옵션 False로 지정
exam.groupby('nclass', as_index = False).agg(mean_math = ('math', 'mean'))
```

```python
# 여러 요약 통계량 한 번에 구하기
exam.groupby('nclass')\
    .agg(mean_math = ('math', 'mean'),
         sum_math = ('math', 'sum'),
         median_math = ('math', 'median'),
         n = ('nclass', 'count'))
```

#### agg()에 자주 사용하는 요약 통계량
#### mean() : 평균
#### std() : 표준편차
#### sum() : 합계
#### median() : 중앙값
#### min() : 최소값
#### max() : 최대값
#### count() : 빈도(개수)

```python
# 모든 변수의 요약 통계량 한 번에 구하기
exam.groupby('nclass').mean()
```

### 집단별로 다시 집단 나누기

```python
mpg = pd.read_csv('mpg.csv')
mpg.groupby(['manufacturer', 'drv']).agg(mean_cty = ('cty', 'mean'))
```

```python
# value_counts()로 집단별 빈도 간단하게 구하기
mpg.groupby('drv').agg(n = ('drv', 'count'))
mpg['drv'].value_counts()
```

### pandas 함수 조합하기
#### 변수 추출 : query()
#### 파생변수 생성 : assign()
#### 변수별로 분리 : groupby()
#### 변수별 함수 적용 : agg()
#### 정렬하기 : sort_values()
#### 특정 범위 출력 : head()

```python
# 제조 회사별로 'suv' 자동차의 도시 및 고속도로 합산 연비 평균을 구해 내림차순으로 정렬하고, 1~5위까지 출력하기
mpg.query('category == "suv"')\
   .assign(total = (mpg['cty'] + mpg['hwy']) / 2)\
   .groupby('manufacturer')\
   .agg(mean_tot = ('total', 'mean'))\
   .sort_values('mean_tot', ascending = False)\
   .head()
```

## 7. 데이터 합치기

### 가로로 합치기
### merge() 이용
#### merge(결합할 데이터 프레임명 나열, how = '결합 방법', on = '기준으로 삼을 변수명')

```python
# 중간고사 데이터 만들기
test1 = pd.DataFrame({'id' : [1, 2, 3, 4, 5],
                       'midterm' : [60, 80, 70, 90, 85]})

# 기말고사 데이터 만들기
test2 = pd.DataFrame({'id' : [1, 2, 3, 4, 5],
                       'final' : [70, 83, 65, 95, 80]})
```

```python
test1
```

```python
test2
```

```python
# id 기준으로 합쳐서 total에 할당
# how = 'left' : 오른쪽에 입력한 데이터 프레임을 왼쪽 데이터 프레임에 결합
total = pd.merge(test1, test2, how = 'left', on = 'id')
total
```

```python
# 다른 데이터를 활용해 변수 추가하기
name = pd.DataFrame({'nclass' : [1, 2, 3, 4, 5],
                      'teacher' : ['kim', 'lee', 'park', 'choi', 'jung']})
name
```

```python
# nclass 기준으로 합쳐서 exam_new에 할당
exam_new = pd.merge(exam, name, how = 'left', on = 'nclass')
exam_new
```

### 세로로 합치기
### concat() 이용
#### concat([결합할 데이터 프레임명 나열])
#### 결합할 데이터 프레임들의 변수명이 동일해야 함

```python
# 학생 1~5번 시험 데이터 만들기
group_a = pd.DataFrame({'id' : [1, 2, 3, 4, 5],
                         'test' : [60, 80, 70, 90, 85]})

# 학생 6~10번 시험 데이터 만들기
group_b = pd.DataFrame({'id' : [6, 7, 8, 9, 10],
                         'test' : [70, 83, 65, 95, 80]})
```

```python
group_a
```

```python
group_b
```

```python
# 데이터 합쳐서 group_all에 할당
group_all = pd.concat([group_a, group_b])
group_all
```

```python
# 인덱스 새로 부여하기
# ignore_index 옵션 True 지정
group_all = pd.concat([group_a, group_b], ignore_index = True)
group_all
```

### midwest.csv : 미국 동북중부 437개 지역의 인구통계 정보

```python
midwest = pd.read_csv('midwest.csv')
midwest
```

### 문제 1 popadults는 해당 지역의 성인 인구, poptotal은 전체 인구를 나타냅니다. midwest 데이터에 '전체 인구 대비 미성연 인구 백분율' 변수를 추가하세요.

```python
midwest = midwest.assign(popminors_percent = (midwest['poptotal'] - midwest['popadults']) / midwest['poptotal'] * 100)
midwest
```

### 문제 2 미성년 인구 백분율이 가장 높은 상위 5개 county(지역)의 미성년 인구 백분율을 출력하세요.

```python
midwest.sort_values('county').head()[['county', 'popminors_percent']]
```

### 문제 3 분류표의 기준에 따라 미성년 비율 등급 변수를 추가하고, 각 등급에 몇 개의 지역이 있는지 알아보세요.

```python
midwest = midwest.assign(grade = np.where(midwest['popminors_percent'] >= 40, 'large', np.where(midwest['popminors_percent'] >= 30, 'middle', 'small')))
```

```python
midwest.groupby('grade').agg(n = ('grade', 'count'))
```

### 문제 4 popasian은 해당 지역의 아시아인 인구를 나타냅니다. '전체 인구 대비 아시아인 인구 백분율' 변수를 추가하고 하위 10개 지역의 state(주), county(지역), 아시아인 인구 백분율을 출력하세요.

```python
midwest.assign(asian_percent = midwest['popasian'] / midwest['poptotal'] * 100)\
       .sort_values('asian_percent')\
       .tail(10)\
       [['state', 'county', 'asian_percent']]
```
