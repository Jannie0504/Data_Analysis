### 요약 통계량 정보별 의미
#### count : 빈도(frequency) - 값의 개수
#### mean : 평균(mean) - 모든 값을 더해 값의 개수로 나눈 값
#### std : 표준편차(standard deviation) - 변수의 값들이 평균에서 떨어진 정도를 나타낸 값
#### min : 최소값(minimum) - 가장 작은 값
#### 25% : 1사분위수(1st quantile) - 하위 25%(4분의 1) 지정에 위치한 값
#### 50% : 중앙값(median) - 하위 50%(중앙) 지점에 위치한 값
#### 75% : 3사분위수(3rd quantile) - 하위 75%(4분의 3) 지점에 위치한 값
#### max : 최대값 - 가장 큰 값

### 조건문을 활용해 파생변수 만들기
#### 전체 자동차 중에서 연비 기준을 충족해 '고연비 합격 판정'을 받은 자동차가 몇 대인지 알아보는 상황
#### 연비가 기준값을 넘으면 합격, 넘지 못하면 불합격
#### 1. 기준값 정하기

```python
mpg['total'].describe()
```

```python
# 그래프 만들기
mpg['total'].plot.hist()

# total 연비의 평균과 중앙값이 약 20
# total 연비가 20~22.5 사이인 자동차 모델이 가장 많음
# total 연비는 대부분 25 이하이고, 25를 넘기는 자동차 모델은 많지 않음
# 연비가 20이 넘는 자동차에 고연비 합격 판정을 내리면 적당할 것으로 판단
```

#### 2. 합격 판정 변수 만들기
#### 조건문 함수 : 조건에 따라 서로 다른 값을 반환하는 함수
#### 조건문 함수를 이용해 특정 조건을 만족했는지 여부에 따라 서로 다른 값을 부여하도록 함
#### numpy 패키지의 where() 이용
#### numpy 패키지 : 배열 연산, 통계치 계산 등 수치 연산을 할 때 자주 사용하는 패키지

```python
import numpy as np
```

```python
# np.where(조건, 조건에 맞을 때 부여하는 값, 조건에 맞지 않을 때 부여하는 값)
mpg['test'] = np.where(mpg['total'] >= 20, 'pass', 'fail')
mpg.head()
```

### 중첩 조건문 활용하기
#### total이 30 이상이면 A, 20~29는 B, 20 미만이면 C 등급으로 분류
#### np.where() 안에 다시 np.where()를 넣는 형식으로 조건문을 중첩
#### 중첩 조건문 : 조건문을 여러 개 중첩한 코드

#### 1. 연비 등급 변수 만들기

```python
# 30 이상이면 A, 아닐 때는 다시 조건문
# 아닐 때 20 이상이면 (30 이상은 아니지만 20 이상이면) B, 그것도 아니면 C
mpg['grade'] = np.where(mpg['total'] >= 30, 'A', np.where(mpg['total'] >= 20, 'B', 'C'))
mpg.head()
```

#### 2. 빈도표와 막대 그래프로 연비 등급 살펴보기

```python
count_grade = mpg['grade'].value_counts()
count_grade
```

```python
count_grade.plot.bar(rot = 0)
```

```python
# 알파벳순으로 막대 정렬하기
# sort_index() 추가 (인덱스 기준으로 오름차순 정렬)
count_grade = mpg['grade'].value_counts().sort_index()
count_grade
```

```python
count_grade.plot.bar(rot = 0)
```

```python
# 필요한 만큼 범주 만들기
# A, B, C, D 등급 변수 만들기
mpg['grade2'] = np.where(mpg['total'] >= 30, 'A', np.where(mpg['total'] >= 25, 'B', np.where(mpg['total'] >= 20, 'C', 'D')))
count_grade2 = mpg['grade2'].value_counts().sort_index()
count_grade2.plot.bar(rot = 0)
```

### 목록에 대한하는 행으로 변수 만들기

```python
# category가 'compact' or 'subcompact' or '2seater' 이면 'small' 부여, 그렇지 않으면 'large' 부여
mpg['size'] = np.where( (mpg['category'] == 'compact') | (mpg['category'] == 'subcompact') | (mpg['category'] == '2seater'), 'small', 'large')
mpg['size'].value_counts()
```

```python
# isin() 이용
# isin([조건1, 조건2, 조건3, ...])
mpg['size'] = np.where(mpg['category'].isin(['compact', 'subcompact', '2seater']), 'small', 'large')
mpg['size'].value_counts()
```
